```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer, PorterStemmer
from wordcloud import WordCloud, STOPWORDS
from textblob import TextBlob
import seaborn as sns
color = sns.color_palette()
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.express as px
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
```

```python
df.head()
```

Remove rows that do not have a review or title

```python
df.dropna(subset = ["Review Text"], inplace=True)
```

```python
df.dropna(subset = ["Title"], inplace=True)
```

```python
df.head()
```

Look at how many ratings there are

```python
fig = px.histogram(df, x="Rating")
fig.update_traces(marker_color="turquoise",marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5)
fig.update_layout(title_text='Product Score')
fig.show()
```

```python
df['Review Text'] = df['Review Text'].astype(str)
```

Make lower case

```python
df['Review Text'] = df['Review Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df['Review Text'][2]
```

Remove punctuation

```python
df['Review Text'] = df['Review Text'].str.replace(r'[^\w\s]+', '')
```

```python
df['Review Text'][2]
```

Remove stopwords

```python
stop = stopwords.words('english')
df['Review Text'] = df['Review Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
df['Review Text'][2]
```

Stemming

```python
ps = PorterStemmer()
```

```python
df['Review Text'] = df['Review Text'].apply(lambda x: " ".join([ps.stem(word) for word in x.split()]))
```

```python
df['Review Text'][2]
```

Adding Sentiment Column

```python
df = df[df['Rating'] != 3]
df['sentiment'] = df['Rating'].apply(lambda rating : +1 if rating > 3 else -1)
```

```python
df.head()
```

```python
positive = df[df['sentiment'] == 1]
negative = df[df['sentiment'] == -1]
```

Creating Training and Testing Data

```python
import numpy as np
```

```python
index = df.index
df['random_number'] = np.random.randn(len(index))
train = df[df['random_number'] <= 0.8]
test = df[df['random_number'] > 0.8]
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
train_matrix = vectorizer.fit_transform(train['Review Text'])
test_matrix = vectorizer.transform(test['Review Text'])
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
X_train = train_matrix
X_test = test_matrix
y_train = train['sentiment']
y_test = test['sentiment']
```

```python
lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
new = np.asarray(y_test)
x = confusion_matrix(predictions,y_test)
x
```

```python
print(classification_report(predictions,y_test))
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(x,cmap= "coolwarm", linecolor = 'black' , linewidth = 1 , annot = True, fmt='')
plt.show()
```
