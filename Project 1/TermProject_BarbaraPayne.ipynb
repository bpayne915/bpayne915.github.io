```python
import thinkstats2
```

```python
firstHalf = df[df.DAY_OF_MONTH < 15]
secondHalf = df[df.DAY_OF_MONTH > 15]
```

```python
distance1PMF = thinkstats2.Pmf(firstHalf.DISTANCE)
distance2PMF = thinkstats2.Pmf(secondHalf.DISTANCE)
```

```python
plt.rcParams["figure.figsize"]=10,10
plt.hist([distance1PMF, distance2PMF], density = "True", label = ['First half of month', 'Second half of month'], bins = 20)
plt.legend(loc = 'upper right')
plt.ylabel('Probability')
plt.xlabel('Distance (miles)')
plt.show()
```

```python
earlyWeek = df[df.DAY_OF_WEEK < 3.5]
lateWeek = df[df.DAY_OF_WEEK > 3.5]
```

```python
distanceWeek1PMF = thinkstats2.Pmf(earlyWeek.DISTANCE)
distanceWeek2PMF = thinkstats2.Pmf(lateWeek.DISTANCE)
```

```python
plt.hist([distanceWeek1PMF, distanceWeek2PMF], density = "True", label = ['First half of week', 'Second half of week'], bins = 20)
plt.legend(loc = 'upper right')
plt.ylabel('Probability')
plt.xlabel('Distance (miles)')
plt.show()
```

```python
import thinkplot
cdf = thinkstats2.Cdf(df.DISTANCE)
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Distance (miles)', ylabel='CDF')
```

```python
log_pops = np.log10(df.DISTANCE)
cdf = thinkstats2.Cdf(df.DISTANCE, label='data')
cdf_log = thinkstats2.Cdf(log_pops, label='data')
```

```python
thinkplot.PrePlot(cols=2)

mu, sigma = log_pops.mean(), log_pops.std()
xs, ps = thinkstats2.RenderNormalCdf(mu, sigma, low=0, high=8)
thinkplot.Plot(xs, ps, label='model', color='0.8')

thinkplot.Cdf(cdf_log) 
thinkplot.Config(xlabel='log10 distance',
                 ylabel='CDF', loc='lower right')
```

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
data = df.DAY_OF_MONTH, df.DISTANCE
ht = CorrelationPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
ht.actual, ht.MaxTestStat()
```
